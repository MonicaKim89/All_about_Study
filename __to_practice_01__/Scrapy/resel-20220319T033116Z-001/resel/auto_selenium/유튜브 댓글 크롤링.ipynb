```python
# 필요한 library import ..

from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
df1 = pd.read_csv('./금연정책유튜브URL.csv')
df1.head()
```

```python
url1 = df1['주소'].tolist()
url1
```

```python
df2 = pd.read_csv('./유튜브URL.csv')
df2
```

```python
def isEx(yurl):
    if yurl in url1:
        return 1
    else: return 0
    
df2['isEx'] = df2.apply(lambda x:isEx(x['주소']),axis=1)
df2
```

```python
df = df2[df2['isEx']==0]
df.drop_duplicates(inplace=True)
```

```python
df.drop(['isEx'],axis=1,inplace=True)
df.to_csv('./유튜브URL_중복제거.csv',index=False,encoding='utf-8-sig')
```

```python
# 주의사항
# with open에서 "w"로 파일을 열었을 경우, 기존 파일 내용을 완전히 지우고 새로 쓴다는 의미입니다.
# 따라서 크롤링 중간에 에러가 났고, 에러부분을 건너뛴 후, 
# 기존 tsv파일 밑으로 이어서 작업하고 싶으시다면
# 반드시 아래 with open ~ 부분 4줄을 지우거나, 주석처리한 후 코드를 실행해 주십시오.

# -----------------------------------------------------------------------------------------------------
with open("/content/drive/MyDrive/Caba_colab/dataset/유튜브_댓글테이블.txt", "w", encoding="utf-8-sig") as f:
    f.write('동영상_제목\t댓글_내용\t댓글_작성시간\n')
    
with open("/content/drive/MyDrive/Caba_colab/dataset/유튜브_동영상테이블.txt", "w", encoding="utf-8-sig") as f:
    f.write('동영상_제목\t동영상_게시일\t댓글_수\t조회수\t좋아요\t동영상_설명_본문\t동영상_URL\n')
# -----------------------------------------------------------------------------------------------------

browser = webdriver.Chrome(executable_path="C:\workspace\jupyter_jiyoon\chromedriver\chromedriver.exe")
browser.maximize_window()


linklists_count = -1
for hreflink in linklists:
    
    linklists_count += 1
    
    # 해당 print는 어느 url에서 error가 발생했는지 체크하는 포인트입니다.
    print("#####" + str(linklists_count)+"번째 linklists 인덱스 작업 중 #####\n")
    
    url = "https://www.youtube.com" + hreflink
    browser.get(url)
    time.sleep(5)
    
    body = browser.find_element_by_tag_name('body')

    height = browser.find_element_by_xpath("""/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div""").size['height']
    while True:
        for i in range(1000):
            body.send_keys(Keys.PAGE_DOWN)
        
        time.sleep(1)
        sizecheck = browser.find_element_by_xpath("""/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div""").size['height']

        if height == sizecheck:
            break
        height = sizecheck
    
    videoNameXpath = """//*[@id="container"]/h1/yt-formatted-string"""
    videoName = browser.find_element_by_xpath(videoNameXpath).text
    videoName = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", videoName)
    videoName = re.sub(" {2,}", " ", videoName)
    print("동영상 제목: "+ videoName)
    
    videoDateXpath = """//*[@id="date"]/yt-formatted-string"""
    videoDate = browser.find_element_by_xpath(videoDateXpath).text
    print("게시일: " + videoDate)
    
    commentNumXpath = """//*[@id="count"]/yt-formatted-string"""


    try:
        commentNum = browser.find_element_by_xpath(commentNumXpath).text
       
    except NoSuchElementException as e:
        for i in range(1000):
            body.send_keys(Keys.PAGE_UP)
        time.sleep(1)
        commentNum = browser.find_element_by_xpath(commentNumXpath).text
        
    commentNum = re.sub(",", "", commentNum)
    commentNum = float(commentNum)
    commentNum = int(commentNum)
    print("댓글 수: " + str(commentNum))
    
    viewCountXpath = """//*[@id="count"]/ytd-video-view-count-renderer/span[1]"""
    viewCount = browser.find_element_by_xpath(viewCountXpath).text
    viewCount = re.sub(r"[가-힣|,| ]", "", viewCount)
    print("조회수: " + viewCount)
    
    goodCountXpath = """//*[@id="text"]"""
    goodCount = browser.find_element_by_xpath(goodCountXpath).text
    print("좋아요 수: " + goodCount)

    # 2021년 하반기에 유튜브 정책에 의해서 싫어요 수가 비공개 처리됨.    
    # badCountXpath = """//*[@id="text"]"""
    # badCount = browser.find_element_by_xpath(badCountXpath).text
    # print("싫어요 수: " + badCount)    
    
    videoTextXpath = """//*[@id="description"]/yt-formatted-string/span[1]"""
    videoText = browser.find_element_by_xpath(videoTextXpath).text
    videoText = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", videoText)
    videoText = re.sub(" {2,}", " ", videoText)
    print("본문 내용: " + videoText)
    
    with open("/content/drive/MyDrive/Caba_colab/dataset/유튜브_동영상테이블.txt", "a", encoding="utf-8-sig") as f:
        f.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(videoName, videoDate, commentNum, viewCount, goodCount, videoText, url))
        
    i = 1
    for i in range(1, int(commentNum)):
        try:
            commentXpath = """/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/ytd-comments/ytd-item-section-renderer/div[3]/ytd-comment-thread-renderer[{0}]/ytd-comment-renderer/div[1]/div[2]/ytd-expander/div/yt-formatted-string[2]""".format(i)
            comment = browser.find_element_by_xpath(commentXpath).text
            comment = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", comment)
            comment = re.sub(" {2,}", " ", comment)
            dateXpath = """/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/ytd-comments/ytd-item-section-renderer/div[3]/ytd-comment-thread-renderer[{0}]/ytd-comment-renderer/div[1]/div[2]/div[1]/div[2]/yt-formatted-string/a""".format(i)
            date = browser.find_element_by_xpath(dateXpath).text
            print(comment)
            print(date)
            
            if comment != " ":
                with open("/content/drive/MyDrive/Caba_colab/dataset/유튜브_댓글테이블.txt", "a", encoding="utf-8-sig") as f:
                    f.write('{}\t{}\t{}\n'.format(videoName, comment, date))
                
        except NoSuchElementException as e:
            continue
        except:
            print("other error")
```

```python
# -----------------------------------------------------------------------------------------------------
with open("./중복제거유튜브_댓글테이블.txt", "w", encoding="utf-8-sig") as f:
    f.write('동영상_제목\t댓글_내용\t댓글_작성시간\n')
    
with open("./중복제거유튜브_동영상테이블.txt", "w", encoding="utf-8-sig") as f:
    f.write('동영상_제목\t동영상_게시일\t댓글_수\t조회수\t좋아요\t싫어요\t동영상_설명_본문\t동영상_URL\n')
# -----------------------------------------------------------------------------------------------------

browser = webdriver.Chrome(executable_path="C:\workspace\jupyter_jiyoon\chromedriver\chromedriver.exe")
browser.maximize_window()
linklists_count = -1
for hreflink in linklists:
    linklists_count += 1
    
    # 해당 print는 어느 url에서 error가 발생했는지 체크하는 포인트입니다.
    print("#####" + str(linklists_count)+"번째 linklists 인덱스 작업 중 #####\n")
    
    browser.get(hreflink)
    time.sleep(5)
    
    body = browser.find_element_by_tag_name('body')

    height = browser.find_element_by_xpath("""/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div""").size['height']

    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)

    soup0 = browser.page_source
    soup = BeautifulSoup(soup0,'html.parser')
    
    #댓글을 막아놓은 영상이 있기 때문에 예외처리를 꼭해준다.
    commentNumXpath = """//*[@id="count"]/yt-formatted-string"""
    try:
        commentNum = browser.find_element_by_xpath(commentNumXpath).text
       
    except NoSuchElementException as e:
        for i in range(5):
            body.send_keys(Keys.PAGE_UP)
        time.sleep(1)
        try:
            commentNum = browser.find_element_by_xpath(commentNumXpath).text
        except NoSuchElementException as e:
            print("댓글이 없는 동영상")
            continue

    commentNum = re.sub("[^0-9]", "", str(commentNum))
    commentNum = float(commentNum)
    commentNum = int(commentNum)
    print("댓글 수: " + str(commentNum))


    videoName = soup.select('yt-formatted-string')[8].text
    videoName = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", videoName)
    videoName = re.sub(" {2,}", " ", videoName)
    print("동영상 제목: ", videoName)

    viewCountXpath = """//*[@id="count"]/ytd-video-view-count-renderer/span[1]"""
    viewCount = browser.find_element_by_xpath(viewCountXpath).text
    viewCount = re.sub(r"[가-힣|A-Z|a-z|,| ]", "", viewCount)
    print("조회수: " + viewCount)

    goodCount = soup.select('yt-formatted-string')[10].text
#     goodCount = re.sub("[K|k|천]","000",goodCount)
#     goodCount = re.sub("[M|m|만]","000000",goodCount)
    print("좋아요 수: " + goodCount)

    badCount = soup.select('yt-formatted-string')[11].text
    badCount = re.sub("[K|k]","000",badCount)
    badCount = re.sub("[M|m]","000000",badCount)
    print("싫어요 수: " + badCount)

    videoDate = soup.select('yt-formatted-string')[9].text
    print("게시일: " + videoDate)

    try:
        videoText = soup.select('yt-formatted-string')[17].text
        videoText = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", videoText)
        videoText = re.sub(" {2,}", " ", videoText)
        print("본문 내용: " + videoText)

    except:
        videoText = ''


    with open("./중복제거유튜브_동영상테이블.txt", "a", encoding="utf-8-sig") as f:
        f.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(videoName, videoDate, commentNum, viewCount, goodCount, badCount, videoText, hreflink))
    
    try:
        i = 0
        while True:
            print(i)
            startP = len(soup.select('yt-formatted-string'))
            if i%50 == 0:
                body.send_keys(Keys.PAGE_DOWN)
                time.sleep(1)
            if i%1000 == 0:
                soup = BeautifulSoup(browser.page_source,'html.parser')
                endP = len(soup.select('yt-formatted-string'))
                if startP == endP:
                    break
            i += 1
        for j in range(endP):
            if j + 2 < endP:
                if soup.select('yt-formatted-string')[j + 2].text == '답글':
                    comment = soup.select('yt-formatted-string')[j].text
                    comment = re.sub(r"[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣|A-Z|a-z|0-9| ]", " ", comment)
                    comment = re.sub(" {2,}", " ", comment)
                    date = soup.select('yt-formatted-string')[j + 1].text
                    print(comment)
                    print(date)
                    if comment != " ":
                        with open("./중복제거유튜브_댓글테이블.txt", "a", encoding="utf-8-sig") as f:
                            f.write('{}\t{}\t{}\n'.format(videoName, comment, date))


    except NoSuchElementException as e:
        continue
    except:
        print("other error")
```
